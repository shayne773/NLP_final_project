```python
#ROUGE-1: ~0.4–0.6+ is good.
#ROUGE-2: ~0.2–0.4+ is reasonable.
#ROUGE-L: ~0.2–0.5+ is expected.
#ROUGE-WE-1 : 0.5–0.8+ for good summaries.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_eval(lower, upper, df, ver):
    good = len([a for a in df['GUSUM_F1'] if a >= lower and a <= upper])
    above = len([a for a in df['GUSUM_F1'] if a > upper])
    below = len([a for a in df['GUSUM_F1'] if a < lower])

    note = str(below/len(df))+'\n\n-----\n'+str(lower)+'\n-----\n\n'+str(good/len(df))+'\n\n-----\n'+ str(upper)+'\n-----\n\n'+str(above/len(df))


    plt.axvspan(lower, upper, color='grey', alpha=0.3, label='Highlighted Range')
    plt.hist(df['GUSUM_F1'], bins=15, color='#2B2B2B', alpha=0.7)
    plt.title(f'Histogram of rouge-{ver}')
    plt.xlabel(f'rouge-{ver} F1')
    plt.ylabel('Frequency')
    plt.text(x=1.1, y=3, s = note, fontsize=12, color='black', rotation=0)
    plt.show()
```

```python

df_1 = pd.read_csv('result/result_rouge_1.csv')
get_eval(0.4,0.6,df_1,'1')
```

```python

df_2 = pd.read_csv('result/result_rouge_2.csv')


get_eval(0.2,0.4,df_2,'2')
```

```python
df_L = pd.read_csv('result/result_rouge_L.csv')


get_eval(0.2,0.5,df_L,'L')
```

```python
df_we = pd.read_csv('result/result_rouge_we.csv')


get_eval(0.5,0.8,df_we,'we')
```
